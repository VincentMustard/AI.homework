2\.Define in your own words the terms constraint, backtracking search, arc consistency,backjumping, min-conflicts, and cycle cutset.

See textbook Chapter6 6.16

constraint: The condition need to be satisfied of solutions.

backtracking search: Walking on a path, if it end or is jam, go back and search for another path unknown.

arc consistency: Xi is arc-consistent with respect to another variable Xj if for every value in the current domain Di there is some value in the domain Dj that satisfies the binary constraint on the arc(Xi, X).

backjumping:  a technique that reduces search space, therefore increasing efficiency.

min-conflicts: Given an initial assignment of values to all the variables of a CSP, the algorithm randomly selects a variable from the set of variables with conflicts violating one or more constraints of the CSP. Then it assigns to this variable the value that minimizes the number of conflicts. If there is more than one value with a minimum number of conflicts, it chooses one randomly. This process of random variable selection and min-conflict value assignment is iterated until a solution is found or a pre-selected maximum number of iterations is reached. 

cycle cutset: Choose a subset S of the CSP’s variables such that the constraint graph becomes a tree after removal of S.S is called a cycle cutset.

```python
import random

if __name__ == '__main__':
    print('there are 3 numbers，please Guess what it is.')
    print('The clue I gave is:')
    print('When I say:                     It means:')
    print('error                           The 3 numbers are not in the mystical numbers.')
    print('Only the number is correct      the number is right , but the position is not right..')
    print('Absolutely right                numbers is right and the position also right.')

    numbers = list(range(10))
    random.shuffle(numbers)
    secret = ''
    for i in range(3):
        secret += str(numbers[i])
    print('there are 3 numbers，You have 10 chances to guess it.')
    rounds = 1
    while rounds <= 10:
        guess = ''
        while len(guess) != 3:
            print()
            guess = input('This is the %s time: ' % rounds)

        if guess == secret:
            print('Congratulations!')
            break

        for i in range(len(guess)):
            if guess[i] == secret[i]:
                print('Absolutely right', end=' ')
            elif guess[i] != secret[i] and guess[i] in secret:
                print('Only the number is correct', end=' ')
            else:
                print('error', end=' ')
        print()

        rounds += 1
        if rounds > 10:
            print('You lost the game,the right answer is %s.' % secret)
```

1. Ball Class
    draw is responsible for moving Ball
    Collision detection, bounce, Ball detection of Paddle 
2. Paddle Class
    draw is responsible for moving Paddle
    Collision detection to determine whether or not to continue ,Monitor keyboard events
3. main cycle
    Drawing the Ball and Paddle

```python
from tkinter import *
import time


class Main:
    def __init__(self, top=None):
        top.title("Game")
        top.resizable(False, False)
        top.wm_attributes("-topmost", 1)
        self.top = top

        self.canvas = Canvas(top, width=500, height=500, bd=0, highlightthickness=0)
        self.canvas.pack()
        self.top.update()
        self.paddle = Paddle(self.canvas, 'blue')
        self.ball = Ball(self.canvas, self.paddle, 'red')

    def run(self):
        while True:
            if not self.ball.hit:
                self.ball.draw()
            else:
                return
            self.top.update_idletasks()
            self.top.update()
            time.sleep(0.01)


class Ball:
    def __init__(self, canvas, paddle, color):
        self.canvas = canvas
        self.paddle = paddle
        self.ball_id = canvas.create_oval(10, 10, 25, 25, fill=color)
        self.canvas.move(self.ball_id, 245, 100)
        self.dx = 3
        self.y = -3
        self.height = self.canvas.winfo_height()
        self.width = self.canvas.winfo_width()
        self.hit = False

    def draw(self):
        self.canvas.move(self.ball_id, self.dx, self.y)
        pos = self.canvas.coords(self.ball_id)
        if pos[1] <= 0 or self.hit_paddle(pos):
            self.y *= -1
        if pos[0] <= 0 or pos[2] >= self.width:
            self.dx *= -1
        if pos[3] >= self.height:
            self.hit = True

    def hit_paddle(self, pos):
        paddle = self.canvas.coords(self.paddle.id)
        if pos[2] >= paddle[0] and pos[0] <= paddle[2] and paddle[1] <= pos[3] <= paddle[3]:
                return True
        return False


class Paddle:
    def __init__(self, canvas, color):
        self.canvas = canvas
        self.id = canvas.create_rectangle(0, 0, 100, 10, fill=color)
        self.dx = 0
        self.canvas.move(self.id, 200, 300)
        self.width = self.canvas.winfo_width()
        self.canvas.bind_all("<Key-Left>", self.move_left)
        self.canvas.bind_all("<Key-Right>", self.move_right)

    def move_left(self, _):
        self.dx = -40
        self.move()

    def move_right(self, _):
        self.dx = 40
        self.move()

    def move(self):
        pos = self.canvas.coords(self.id)
        if pos[0] + self.dx >= 0 and pos[2] + self.dx <= self.width:
            self.canvas.move(self.id, self.dx, 0)


if __name__ == "__main__":
    try:
        Main(Tk()).run()
    except:
        pass
```
